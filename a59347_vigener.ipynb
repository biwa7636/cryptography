# Encrypt and Decypt 

```python
def getPos(a):
    # get the position in the alphabet table 
    # A -> 1, Z -> 26
    return ord(a) - ord('A') + 1

def encrypt_letter(target_letter, key_letter):
    # P+T -> J
    # C+W -> Z
    diff = (getPos(target_letter) + getPos(key_letter)) % 26 
    if diff == 0:
        diff = 26
    return chr(ord('A')-1+diff)

def vigener_encrypt(message, key):
    # vigener('whatanicedaytoday', 'crypto') -> ZZZJUCLUDTUNWGCQS
    if not message.isalpha() or not key.isalpha():
        print 'Message and Key have to be alphabet letters ONLY!'
        return
    
    message = message.upper()
    key = key.upper()
    result = ""
    for index, letter_message in enumerate(message): 
        letter_key = key[index % len(key)]
        result += encrypt_letter(letter_message, letter_key)
    return result

#TODO: decryption
```
